###  Target variable

* Class imbalance -  there are a significant amount more customers that will not make a transaction

```python
sns.set()
plt.figure(figsize = (7,7))
sns.countplot(train['target'])
plt.title('Target Variable')
plt.show()
```

```python
plt.figure(figsize = (8,8))
train.target.value_counts().plot(kind = 'pie', autopct = '%1.0f%%')
```

A check for linear combinations

```python
train_df_correlations = train.drop(["target"], axis=1).corr()
train_df_correlations = train_df_correlations.values.flatten()
train_df_correlations = train_df_correlations[train_df_correlations != 1]

test_df_correlations = test.corr()
test_df_correlations = test_df_correlations.values.flatten()
test_df_correlations = test_df_correlations[test_df_correlations != 1]

plt.figure(figsize=(18,7))

sns.distplot(train_df_correlations, color="Blue", label = "train")
sns.distplot(test_df_correlations, color="Red", label = "test")

plt.ylabel("Density")
plt.title("  Feature Correlations"); 
plt.legend();
```

### Feature correlation

* Very little corrrelation between features

```python
features = ['var_0', 'var_1', 'var_2', 'var_3', 'var_4','var_5', 'var_6', 'var_7', 'var_8', 'var_9', 'var_10', 'var_11', 'var_12', 'var_13', 'var_14', 'var_15', 'var_16', 'var_17', 'var_18', 'var_19', 'var_20','var_190', 'var_191', 'var_192', 'var_193', 'var_194', 'var_195','var_196', 'var_197', 'var_198', 'var_199']
```

```python
feature_correlations = train[features].corr().abs().unstack().sort_values(kind = 'quicksort').reset_index()
```

```python
feature_correlations = feature_correlations[feature_correlations['level_0'] != feature_correlations['level_1']]
```
